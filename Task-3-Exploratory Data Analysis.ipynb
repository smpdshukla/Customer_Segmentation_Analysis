```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("Telco_Customer_Churn_Dataset.csv")
```

```python
#Summary Statistics
print("\nMean:")
print(data.mean(numeric_only=True))
print("\nMedian:")
print(data.median(numeric_only=True))
print("\nMode:")
print(data.mode())
print("\nSummary Statistics:")
print(data.describe(include='all'))
```

```python
#Histogram
num_cols = data.select_dtypes(include=["int64","float64"]).columns
for col in num_cols:
    plt.figure(figsize=(6,4))
    plt.hist(data[col],bins=20,edgecolor="black",color="skyblue")
    plt.suptitle(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#Box plot
num_cols = data.select_dtypes(include=["int64","float64"]).columns
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=data[col],color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.ylabel(col)
    plt.show()
    
```

```python
#Churn Proportion
plt.figure(figsize=(6,6))
data['Churn'].value_counts().plot.pie(autopct="%1.1f%%",colors=["lightgreen","salmon"],startangle=90,labels=["No Churn","Churn"])
plt.title("Churn vs Non-Churn Distriution")
plt.ylabel("")
plt.show()
```
